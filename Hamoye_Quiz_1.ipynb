**Importing Numpy and Pandas Library**

```python
import numpy as np
import pandas as pd

```

**Loading Dataset**

```python
df = pd.read_csv("https://github.com/HamoyeHQ/HDSC-Introduction-to-Python-for-machine-learning/files/7768140/FoodBalanceSheets_E_Africa_NOFLAG.csv", encoding = "latin-1")
df
```

**Total Number of the Sum of Processing in 2017**

```python
# Filter the DataFrame for 'Processing' and group by 'Element'
processing_2017_grouped = df[df['Element'] == 'Processing'].groupby('Element')['Y2017'].sum()
print(processing_2017_grouped)
# Access the sum for 'Processing' in 2017
total_processing_2017 = processing_2017_grouped.get('Processing', 0)

print(total_processing_2017)
```

**Area that had the Highest Sum in 2017**

```python
# Select the 'Area' and 'Y2017' columns
selected_columns = df[['Area', 'Y2017']]
selected_columns
```

```python
# Group by 'Area' and calculate the sum of 'y2017' for each area
grouped = selected_columns.groupby('Area')['Y2017'].sum()
grouped
```

```python
# Find the area with the highest value in 'y2017'
area_with_highest_value = grouped.idxmax()

area_with_highest_value
```

**Year with the Highest Sum of Stock Variation**

```python
# Select the Years and Element columns
columns = df.loc[:, ['Element', 'Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']]
columns
```

```python
# Filter for "Stock Variation" in Element
stock_variation_df = columns[columns['Element'] == 'Stock Variation']
stock_variation_df
```

**Which year had the least correlation with "Element Code"?**

```python
# Select relevant columns
corr_columns = df.loc [:, ['Element Code', 'Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']]
corr_columns

```

```python
# Calculate the correlation for each year using "for loop"
correlations = {}
for y_col in corr_columns[1:]:
    correlation = corr_columns['Element Code'].corr(df[y_col])
    correlations[y_col] = correlation
    print (correlations)


```

```python
# Find the year with the least correlation
least_correlated_year = min(correlations, key=correlations.get)
least_correlation_value = correlations[least_correlated_year]
print(least_correlated_year)
```

**Mean and Standard Deviation Across the Dataset for the Year 2017**

```python
# Calculate the mean for the year 2017
mean_2017 = df['Y2017'].mean()

# Calculate the standard deviation for the year 2017
std_deviation_2017 = df['Y2017'].std()

# Round the values to two decimal places
mean_2017 = round(mean_2017, 2)
std_deviation_2017 = round(std_deviation_2017, 2)

print(mean_2017)
print(std_deviation_2017)
```

**Total number and percentage of missing data in 2014 to 3 decimal places**

```python
# Count the number of missing values in the 'Y2014' column
missing_data_count = df['Y2014'].isnull().sum()

# Calculate the total number of rows in the DataFrame
total_rows = len(df)

# Calculate the percentage of missing data in 2014
percentage_missing_data = (missing_data_count / total_rows) * 100

# Round the percentage to three decimal places
percentage_missing_data = round(percentage_missing_data, 3)

print(missing_data_count)
print(percentage_missing_data, "%")
```

**Area that had the 7th Lowest Sum in 2017**

```python
#Select Columns
area_2017 = df.loc[:, ['Area', 'Y2017']]
print(area_2017)
```

```python
# Group by 'Area' and calculate the sum of 'Y2017'
grouped = area_2017.groupby('Area')['Y2017'].sum()

print(grouped)
```

```python
# Sort the grouped data in ascending order and get the 7th lowest sum
seventh_lowest_sum = grouped.sort_values().iloc[6]
print(seventh_lowest_sum)

```

```python
# Find the areas with the 7th lowest sum in 2017
areas_with_seventh_lowest_sum = grouped[grouped == seventh_lowest_sum].index.tolist()

print(areas_with_seventh_lowest_sum)
```
